```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
start_year = '1952'
end_year = '2020'

base_link = 'https://www.weltfussball.de/spielplan/dfb-pokal-'
end_link = '/3/'
links = []


link = 'https://www.weltfussball.de/spielplan/dfb-pokal-2020-2021-1-runde/0/'

for year in range(int(start_year), 1957):
    link_1 = base_link + str(year) + '-'+ str(year+1) + '-1-runde'+ end_link
    link_2 = base_link + str(year) + '-'+ str(year+1) + '-2-runde'+ end_link
    link_3 = base_link + str(year) + '-'+ str(year+1) + '-achtelfinale'+ end_link
    link_4 = base_link + str(year) + '-'+ str(year+1) + '-viertelfinale'+ end_link
    link_5 = base_link + str(year) + '-'+ str(year+1) + '-halbfinale'+ end_link
    link_6 = base_link + str(year) + '-'+ str(year+1) + '-endspiel'+ end_link
    
    #links.append([link_1, link_2])
    links.append(link_1)
    links.append(link_2)
    links.append(link_3)
    links.append(link_4)
    links.append(link_5)
    links.append(link_6)
```

```python
for link in links:
    html = requests.get(link).text
    soup = bs(html, 'html.parser')

    # get whole table
    table = soup.find('table', class_='standard_tabelle')

    # tablerow
    tr = table.find_all('tr')

    date = ''

    for r in tr:
        try:
            temp_date = r.find('td', nowrap='nowrap').text
            if temp_date != '':
                date = temp_date
            #print(date.text)
        except:
            pass

        try:
            match = r.find_all('a')
            home = match[0].text
            away = match[1].text
            result = match[2].text
        except:
            pass

        print(f'{date}\t  {home} \t \t \t {away} \t \t \t {result}')

```
