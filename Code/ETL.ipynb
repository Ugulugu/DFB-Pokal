```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
start_year = '1952'
end_year = '2020'

base_link = 'https://www.weltfussball.de/spielplan/dfb-pokal-'
end_link = '/3/'

links = []
seasons = []


for year in range(int(start_year), 1955):
    season_dic = {
        'SeasonStart': '',
        'SeasonEnd': '',
        

        'DFB_link1':'',
        'DFB_link2':'',
        'DFB_link3':'',
        'DFB_link4':'',
        'DFB_link5':'',
        'DFB_link6':''
    }
    
    season_dic['SeasonStart'] = year
    season_dic['SeasonEnd'] = year + 1
    
    season_dic['DFB_link1'] = base_link + str(year) + '-'+ str(year+1) + '-1-runde'+ end_link
    season_dic['DFB_link2'] = base_link + str(year) + '-'+ str(year+1) + '-2-runde'+ end_link
    season_dic['DFB_link3'] = base_link + str(year) + '-'+ str(year+1) + '-achtelfinale'+ end_link
    season_dic['DFB_link4'] = base_link + str(year) + '-'+ str(year+1) + '-viertelfinale'+ end_link
    season_dic['DFB_link5'] = base_link + str(year) + '-'+ str(year+1) + '-halbfinale'+ end_link
    season_dic['DFB_link6'] = base_link + str(year) + '-'+ str(year+1) + '-endspiel'+ end_link
    
    seasons.append(season_dic)
    #print(season_dic['SeasonStart'])
    #print(season_dic['SeasonEnd'])
    #print()
    
for season in seasons:
    print(season)
    print()
```

```python
data = pd.DataFrame({'Date': [], 'Leg': [], 'Home':[], 'Away': [], 'Result':[]})

for season in seasons:
    links = [season['DFB_link1'],season['DFB_link2'],season['DFB_link3'],season['DFB_link4'],season['DFB_link5'],season['DFB_link6']]
    
    for link in links:
        html = requests.get(link).text
        soup = bs(html, 'html.parser')

        # get whole table
        table = soup.find('table', class_='standard_tabelle')

        # tablerow
        tr = table.find_all('tr')

        date = ''

        for r in tr:
            try:
                temp_date = r.find('td', nowrap='nowrap').text
                if temp_date != '':
                    date = temp_date
                #print(date.text)
            except:
                pass

            try:
                leg = soup.title.text
                match = r.find_all('a')
                home = match[0].text
                away = match[1].text
                result = match[2].text
            except:
                pass

            temp_df = pd.DataFrame({'Date': [date], 'Leg': [leg], 'Home':[home], 'Away': [away], 'Result':[result]})
            data = data.append(temp_df)
            #print(f'{date},{home},{away},{result}')

print('Fertig')
```
