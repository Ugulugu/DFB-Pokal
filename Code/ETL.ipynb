```python
data = pd.DataFrame({'Date': [], 'Home':[], 'Away': [], 'Result':[]})
```

```python
for link in links:
    html = requests.get(link).text
    soup = bs(html, 'html.parser')

    # get whole table
    table = soup.find('table', class_='standard_tabelle')

    # tablerow
    tr = table.find_all('tr')

    date = ''

    for r in tr:
        try:
            temp_date = r.find('td', nowrap='nowrap').text
            if temp_date != '':
                date = temp_date
            #print(date.text)
        except:
            pass

        try:
            match = r.find_all('a')
            home = match[0].text
            away = match[1].text
            result = match[2].text
        except:
            pass
        
        temp_df = pd.DataFrame({'Date': [date], 'Home':[home], 'Away': [away], 'Result':[result]})
        data = data.append(temp_df)
        #print(f'{date},{home},{away},{result}')

print('Fertig')
```
